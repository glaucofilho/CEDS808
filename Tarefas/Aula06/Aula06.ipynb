```python
# ........................REGRESSAO LINEAR ....................................
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Gerando dados sintéticos
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1  # Adicionando ruído aos dados

# Criando e treinando o modelo de regressão linear
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)

# Fazendo previsões
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_pred = linear_regressor.predict(X_test)

# Calculando o MSE
mse = mean_squared_error(y, linear_regressor.predict(X))
# Calculando o R^2
r2 = r2_score(y, linear_regressor.predict(X))

# Plotando os dados e as previsões da regressão linear
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Dados de Treinamento')
plt.plot(X_test, y_pred, color='red', linestyle='--', label='Previsões')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Exemplo de Regressão Linear')
plt.legend()
plt.grid(True)
plt.show()

print(f"MSE: {mse:.2f}")
print(f"R^2: {r2:.2f}")
```

```python
# ........................KNN REGRESSOR ....................................
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Gerando dados sintéticos
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

# Adicionando ruído aos dados
y[::5] += 1 * (0.5 - np.random.rand(8))

# Criando o modelo de regressão k-NN
knn_regressor = KNeighborsRegressor(n_neighbors=5)

# Ajustando o modelo aos dados
knn_regressor.fit(X, y)

# Gerando valores de teste
X_test = np.linspace(0, 5, 500)[:, np.newaxis]

# Fazendo previsões para os valores de teste
y_pred = knn_regressor.predict(X_test)

# Plotando os resultados
plt.scatter(X, y, color='darkorange', label='Dados de treinamento')
plt.plot(X_test, y_pred, color='navy', label='Regressão k-NN')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Regressão k-NN')
plt.legend()
plt.show()
```

```python
# ........................ARVORE DE DECISAO REGRESSOR ....................................
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Gerando dados sintéticos
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1  # Adicionando ruído aos dados

# Criando e treinando o modelo de regressão com árvore de decisão
tree_regressor = DecisionTreeRegressor(max_depth=5)
tree_regressor.fit(X, y)

# Dados para fazer previsões (mais densamente espaçados)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

# Fazendo previsões
y_pred = tree_regressor.predict(X_test)

# Plotando os dados e as previsões da árvore de decisão
plt.figure(figsize=(10, 6))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="Dados de Treinamento")
plt.plot(X_test, y_pred, color="cornflowerblue", label="Previsões da Árvore de Decisão", linewidth=2)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Exemplo de Regressão com Árvore de Decisão")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Gerando dados sintéticos
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1  # Adicionando ruído aos dados

# Valores para variar o parâmetro max_depth
max_depths = [2, 3, 5, 7, 9]

# Número de linhas e colunas para a matriz de subplots
n_plots = len(max_depths)

# Plotagem dos resultados para diferentes valores de max_depth
plt.figure(figsize=(12, 6 * n_plots))

for i, max_depth in enumerate(max_depths, start=1):
    # Criando e treinando o modelo de regressão com árvore de decisão
    tree_regressor = DecisionTreeRegressor(max_depth=max_depth)
    tree_regressor.fit(X, y)

    # Dados para fazer previsões (mais densamente espaçados)
    X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
    y_pred = tree_regressor.predict(X_test)

    # Calculando o MSE
    mse = mean_squared_error(y, tree_regressor.predict(X))
    # Calculando o R^2
    r2 = r2_score(y, tree_regressor.predict(X))

    # Plotando os resultados
    plt.subplot(n_plots, 1, i)
    plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="Dados de Treinamento")
    plt.plot(X_test, y_pred, color="cornflowerblue", label="Previsões")
    plt.xlabel("Feature")
    plt.ylabel("Target")
    plt.title(f"max_depth={max_depth}\nMSE={mse:.2f}\nR^2={r2:.2f}")
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plotando a árvore de decisão
plt.figure(figsize=(14, 10))
plot_tree(tree_regressor, filled=True, feature_names=["Feature"], rounded=True)
plt.title("Árvore de Decisão para Regressão")
plt.show()

# Acessar o nó raiz (para impressão do valor)
root_value = tree_regressor.tree_.value[0]

# Print do valor do nó raiz
print("Valor do Nó Raiz:", root_value)

# Acessar os índices dos nós abaixo do nó raiz
left_child_index = tree_regressor.tree_.children_left[0]
right_child_index = tree_regressor.tree_.children_right[0]

# Acessar os valores dos nós abaixo do nó raiz
left_child_value = tree_regressor.tree_.value[left_child_index]
right_child_value = tree_regressor.tree_.value[right_child_index]

# Print dos valores dos nós abaixo do nó raiz
print("Valor do Nó à Esquerda do Raiz:", left_child_value)
print("Valor do Nó à Direita do Raiz:", right_child_value)
```

```python

# ........................EXEMPLO SIMPLÓRIO REPLICANDO PASSOS DO SVR ....................................
import numpy as np
from scipy.optimize import minimize

# Dados de treinamento
X_train = np.array([[50], [70], [90], [110], [130], [150], [170], [190], [210], [230]])  # Área (m²)
y_train = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550])                   # Preço ($1000)

# Parâmetros do modelo SVM regressor
C = 1  # Parâmetro de regularização
epsilon = 0.1  # Margem de erro

# Cálculo do vetor de pesos (w) e viés (b)
n_samples, n_features = X_train.shape
w = np.zeros(n_features)
b = 0

# Função de perda encapsulada com dados de treinamento
def loss_function_closure(X, y):
    def loss_function(params):
        w = params[:-1]
        b = params[-1]
        error = np.maximum(0, np.abs(X.dot(w) + b - y) - epsilon)
        return 0.5 * np.sum(w ** 2) + C * np.sum(error)
    return loss_function

# Otimização da função de perda para encontrar w e b
loss_function = loss_function_closure(X_train, y_train)
res = minimize(loss_function, np.hstack((w, b)), method='L-BFGS-B')
params = res.x
w = params[:-1]
b = params[-1]

area_casa = 80
preco_previsto = area_casa * w + b
print("Preço previsto para uma casa com área de 80 m²:", preco_previsto)
```

```python
from sklearn.svm import NuSVR
import numpy as np

# Dados de treinamento
X_train = np.array([[50], [70], [90], [110], [130], [150], [170], [190], [210], [230]])  # Área (m²)
y_train = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550])                      # Preço ($1000)

# Valores de nu a serem testados
nu_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Para cada valor de nu, criamos e treinamos um modelo NuSVR e imprimimos o preço previsto para uma casa com área de 80 m²
for nu in nu_values:
    # Criando e treinando o modelo NuSVR
    nusvr_regressor = NuSVR(kernel='linear', nu=nu)
    nusvr_regressor.fit(X_train, y_train)

    # Prevendo o preço de uma casa com área de 80 m²
    area_casa = np.array([[80]])  # Área da casa a ser prevista
    preco_previsto = nusvr_regressor.predict(area_casa)

    # Imprimindo o preço previsto para o valor atual de nu
    print(f"Preço previsto para uma casa com área de 80 m² (nu={nu}):", preco_previsto[0])
```

```python
# ......................................MLP REGRESSOR ...................................................
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# Dados de treinamento
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([2, 4, 6, 8, 10])  # A relação é y = 2 * X

# Criando e treinando o modelo MLP Regressor
mlp_regressor = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_regressor.fit(X_train, y_train)

# Dados para previsão
X_test = np.array([[6], [7], [8]])

# Realizando as previsões
predictions = mlp_regressor.predict(X_test)

# Plotando os resultados
plt.scatter(X_train, y_train, color='blue', label='Dados de Treinamento')
plt.plot(X_test, predictions, color='red', linestyle='--', marker='o', label='Previsões')
plt.xlabel('X')
plt.ylabel('y')
plt.title('MLP Regressor Example')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# Gerando dados de treinamento
X_train = np.linspace(0, 10, 100).reshape(-1, 1)
y_train = np.sin(X_train).ravel() + np.random.normal(0, 0.1, size=(100,))

# Criando e treinando o modelo MLP Regressor
mlp_regressor = MLPRegressor(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_regressor.fit(X_train, y_train)

# Gerando dados de teste
X_test = np.linspace(0, 10, 200).reshape(-1, 1)

# Realizando previsões
predictions = mlp_regressor.predict(X_test)

# Calculando o erro
mse = np.mean((predictions - np.sin(X_test).ravel()) ** 2)
print("Erro médio quadrático (MSE):", mse)

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Dados de Treinamento')
plt.plot(X_test, predictions, color='red', label='Previsões')
plt.xlabel('X')
plt.ylabel('y')
plt.title('MLP Regressor Example with More Data')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# Gerando dados de treinamento
X_train = np.linspace(0, 10, 100).reshape(-1, 1)
y_train = np.sin(X_train).ravel() + np.random.normal(0, 0.1, size=(100,))

# Definindo os parâmetros da rede neural
hidden_layer_sizes = [(100,), (50,), (50, 20)]
activations = ['relu', 'tanh']
solvers = ['adam', 'sgd']
max_iters = [500, 1000]

# Variando os parâmetros
for hidden_layer_size in hidden_layer_sizes:
    for activation in activations:
        for solver in solvers:
            for max_iter in max_iters:
                # Criando e treinando o modelo MLP Regressor
                mlp_regressor = MLPRegressor(hidden_layer_sizes=hidden_layer_size, activation=activation, solver=solver, max_iter=max_iter, random_state=42)
                mlp_regressor.fit(X_train, y_train)

                # Gerando dados de teste
                X_test = np.linspace(0, 10, 200).reshape(-1, 1)

                # Realizando previsões
                predictions = mlp_regressor.predict(X_test)

                # Calculando o erro
                mse = np.mean((predictions - np.sin(X_test).ravel()) ** 2)

                # Imprimindo os parâmetros e o erro
                print(f"Parâmetros: hidden_layer_sizes={hidden_layer_size}, activation={activation}, solver={solver}, max_iter={max_iter}")
                print("Erro médio quadrático (MSE):", mse)

                # Plotando os resultados
                plt.figure(figsize=(10, 6))
                plt.scatter(X_train, y_train, color='blue', label='Dados de Treinamento')
                plt.plot(X_test, predictions, color='red', label='Previsões')
                plt.xlabel('X')
                plt.ylabel('y')
                plt.title(f'MLP Regressor Example with Parameters: {hidden_layer_size}, {activation}, {solver}, {max_iter}')
                plt.legend()
                plt.grid(True)
                plt.show()
```
